```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import f1_score

df = pd.read_csv(r'pima-indians-diabetes.data', sep=',')
print(df.shape)
df.head()
```

```python
#导入相关包
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
#计算特征相关性并可视化
corr_matrix = df.corr(method='spearman') # pearson 皮尔逊, 是spearman 斯皮尔曼 
plt.figure(figsize=(25, 15))
sns.heatmap(corr_matrix, annot= True)
```

```python
from collections import Counter
x_cols = [col for col in df.columns if col != 'Outcome']
y_col = 'Outcome'
X = df[x_cols].values
y = df[y_col].values

X_train, X_test, y_train, y_test = train_test_split(
    df[x_cols], df[y_col], test_size=0.1, random_state=42, shuffle=True, stratify=df[y_col]
)
print('Distribution of y_train {}'.format(Counter(y_train)))
print('Distribution of y_test {}'.format(Counter(y_test)))
```

### 逻辑回归模型

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler

# 标准化数据
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# 增加 max_iter 并使用标准化后的数据
lr_model = LogisticRegression(max_iter=1000)
cv_score = cross_val_score(lr_model, X_train_scaled, y_train, scoring='f1', cv=5)
lr_model.fit(X_train, y_train)  # 训练 Logistic Regression 模型
print('Logistic Regression F1 Score (CV):', cv_score.mean())
```

### 随机森林模型

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# 初始化随机森林模型
rf_model = RandomForestClassifier(random_state=42)

# 定义超参数网格
rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}

# 使用 GridSearchCV 进行网格搜索
rf_grid = GridSearchCV(rf_model, rf_params, cv=5, scoring='f1')
rf_grid.fit(X_train, y_train)

# 输出最佳 F1 分数和最佳参数
print('Random Forest Best F1 Score:', rf_grid.best_score_)
print('Random Forest Best Parameters:', rf_grid.best_params_)
```

### 支持向量机模型

```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"  # 解决可能的库冲突问题
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, train_test_split

# 检查数据集大小
print(X_train.shape, y_train.shape)

# 使用数据子集（可选）
X_train_small, _, y_train_small, _ = train_test_split(X_train, y_train, train_size=0.1, random_state=42)

# 初始化 SVM 模型
svc_model = SVC(random_state=42, max_iter=1000, tol=1e-3, probability=True)  # 启用概率预测

# 定义超参数网格
svc_params = {
    'C': [0.1, 1],  # 减少 C 的取值
    'kernel': ['linear']  # 只使用线性核
}

# 使用 GridSearchCV 进行网格搜索
svc_grid = GridSearchCV(
    svc_model,
    svc_params,
    cv=3,  # 减少交叉验证折数
    scoring='f1',
    n_jobs=1  # 关闭并行计算
)

# 训练模型
svc_grid.fit(X_train_small, y_train_small)  # 使用数据子集

# 输出最佳 F1 分数和最佳参数
print('SVM Best F1 Score:', svc_grid.best_score_)
print('SVM Best Parameters:', svc_grid.best_params_)
```

### K近邻模型

```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier()
knn_params = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance']
}
knn_grid = GridSearchCV(knn_model, knn_params, cv=5, scoring='f1')
knn_grid.fit(X_train, y_train)
print('KNN Best F1 Score:', knn_grid.best_score_)
print('KNN Best Parameters:', knn_grid.best_params_)
```

### XGBoost模型

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

# 初始化 XGBoost 模型
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')

# 定义超参数网格
xgb_params = {
    'n_estimators': [100, 200],  # 树的数量
    'max_depth': [3, 5, 7],      # 树的最大深度
    'learning_rate': [0.01, 0.1], # 学习率
    'subsample': [0.8, 1.0]      # 样本子集比例
}

# 使用 GridSearchCV 进行网格搜索
xgb_grid = GridSearchCV(xgb_model, xgb_params, cv=5, scoring='f1')
xgb_grid.fit(X_train, y_train)

# 输出最佳 F1 分数和最佳参数
print('XGBoost Best F1 Score:', xgb_grid.best_score_)
print('XGBoost Best Parameters:', xgb_grid.best_params_)
```

### 为每个模型绘制混淆矩阵（分开绘制）

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Predicted 0', 'Predicted 1'], 
                yticklabels=['Actual 0', 'Actual 1'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()

# 为每个模型绘制混淆矩阵
plot_confusion_matrix(y_test, lr_model.predict(X_test), 'Confusion Matrix - Logistic Regression')
plot_confusion_matrix(y_test, rf_grid.best_estimator_.predict(X_test), 'Confusion Matrix - Random Forest')
plot_confusion_matrix(y_test, svc_grid.best_estimator_.predict(X_test), 'Confusion Matrix - SVM')
plot_confusion_matrix(y_test, knn_grid.best_estimator_.predict(X_test), 'Confusion Matrix - KNN')
plot_confusion_matrix(y_test, xgb_grid.best_estimator_.predict(X_test), 'Confusion Matrix - XGBoost')
```

### 绘制多个模型的 ROC 曲线

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# 定义函数绘制 ROC 曲线
def plot_roc_curve_multiple(y_true, models, model_names, X_test):
    plt.figure(figsize=(8, 6))
    for model, name in zip(models, model_names):
        y_pred_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_true, y_pred_prob)
        auc_score = roc_auc_score(y_true, y_pred_prob)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {auc_score:.2f})')
    
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guess')
    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR)')
    plt.title('ROC Curve - Model Comparison')
    plt.legend()
    plt.show()

# 模型列表
models = [
    lr_model,
    rf_grid.best_estimator_,
    svc_grid.best_estimator_,
    knn_grid.best_estimator_,
    xgb_grid.best_estimator_
]
model_names = ['Logistic Regression', 'Random Forest', 'SVM', 'KNN', 'XGBoost']

# 绘制 ROC 曲线
plot_roc_curve_multiple(y_test, models, model_names, X_test)
```

### 多个模型的分类报告

```python
from sklearn.metrics import classification_report
from prettytable import PrettyTable

def generate_classification_report_table(y_true, models, model_names, X_test):
    reports = []
    for model, name in zip(models, model_names):
        y_pred = model.predict(X_test)
        report = classification_report(y_true, y_pred, output_dict=True)
        report_df = pd.DataFrame(report).transpose()
        report_df['Model'] = name
        reports.append(report_df)
    
    # 合并所有报告
    final_report = pd.concat(reports)
    
    # 调整格式
    final_report = final_report.reset_index().set_index(['Model', 'index'])  # 将 Model 和 index 作为多级索引
    final_report = final_report[['precision', 'recall', 'f1-score', 'support']]  # 只保留关键指标
    final_report = final_report.round(2)  # 将数值格式化为两位小数
    
    # 过滤掉 accuracy、macro avg 和 weighted avg
    final_report = final_report.loc[(slice(None), ['0', '1']), :]
    
    return final_report

# 生成分类报告表格
report_table = generate_classification_report_table(y_test, models, model_names, X_test)

# 创建两个表格，分别对比 Class 0 和 Class 1
table_class_0 = PrettyTable()
table_class_0.field_names = ['Model', 'Class', 'Precision', 'Recall', 'F1-Score', 'Support']

table_class_1 = PrettyTable()
table_class_1.field_names = ['Model', 'Class', 'Precision', 'Recall', 'F1-Score', 'Support']

# 填充表格
for (model, cls), row in report_table.iterrows():
    if cls == '0':
        table_class_0.add_row([model, cls, row['precision'], row['recall'], row['f1-score'], row['support']])
    elif cls == '1':
        table_class_1.add_row([model, cls, row['precision'], row['recall'], row['f1-score'], row['support']])

# 打印表格
print("Class 0 Comparison:")
print(table_class_0)
print("\nClass 1 Comparison:")
print(table_class_1)
```

### 绘制多个模型的精确率-召回率曲线

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# 定义函数绘制精确率-召回率曲线
def plot_precision_recall_curve_multiple(y_true, models, model_names, X_test):
    plt.figure(figsize=(8, 6))
    for model, name in zip(models, model_names):
        y_pred_prob = model.predict_proba(X_test)[:, 1]
        precision, recall, _ = precision_recall_curve(y_true, y_pred_prob)
        plt.plot(recall, precision, label=f'{name}')
    
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve - Model Comparison')
    plt.legend()
    plt.show()

# 绘制精确率-召回率曲线
plot_precision_recall_curve_multiple(y_test, models, model_names, X_test)
```

### 综合上述分析，随机森林和XGBoost模型的整体效果是最好的，各方面的实验结果都好于其他的模型。
但是从上述结果也能看出Class = 1时的指标较低，下面我们组将通过**过采样**的方法来提高这一指标。

### 采用过采样

```python
from imblearn.over_sampling import SMOTE

# 初始化 SMOTE
smote = SMOTE(random_state=42)

# 对训练数据进行过采样
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# 查看过采样后的类别分布
print('Resampled Class Distribution:', Counter(y_train_resampled))
```

### 重新训练

```python
# 随机森林
rf_model_resampled = RandomForestClassifier(random_state=42)
rf_model_resampled.fit(X_train_resampled, y_train_resampled)

# XGBoost
xgb_model_resampled = XGBClassifier(random_state=42)
xgb_model_resampled.fit(X_train_resampled, y_train_resampled)
```

```python
# 随机森林
y_pred_rf = rf_model_resampled.predict(X_test)

# XGBoost
y_pred_xgb = xgb_model_resampled.predict(X_test)
```

### 绘制混淆矩阵

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Predicted 0', 'Predicted 1'], 
                yticklabels=['Actual 0', 'Actual 1'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()

# 随机森林
plot_confusion_matrix(y_test, y_pred_rf, 'Confusion Matrix - Random Forest (Resampled)')

# XGBoost
plot_confusion_matrix(y_test, y_pred_xgb, 'Confusion Matrix - XGBoost (Resampled)')
```

### 绘制ROC曲线

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

def plot_roc_curve_multiple(y_true, models, model_names, X_test):
    plt.figure(figsize=(8, 6))
    for model, name in zip(models, model_names):
        y_pred_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_true, y_pred_prob)
        auc_score = roc_auc_score(y_true, y_pred_prob)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {auc_score:.2f})')
    
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guess')
    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR)')
    plt.title('ROC Curve - Model Comparison')
    plt.legend()
    plt.show()

# 模型列表
models = [rf_model_resampled, xgb_model_resampled]
model_names = ['Random Forest (Resampled)', 'XGBoost (Resampled)']

# 绘制 ROC 曲线
plot_roc_curve_multiple(y_test, models, model_names, X_test)
```
