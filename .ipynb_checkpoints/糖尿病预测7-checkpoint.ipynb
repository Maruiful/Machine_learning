### 步骤5 对输入特征进行降维，选择PCA。

首先使用StandardScaler对输入特征X进行了标准化处理。标准化是降维技术中的一个重要步骤，它能确保每个特征对结果的影响是平等的。然后创建一个PCA对象，并指定n_components=0.99以保留99%的信息。最后，对标准化后的数据X_std应用PCA进行降维，得到新的特征矩阵X_pca。  
输入特征（X）：包含 8 个原始特征（如Glucose血糖、BMI体质指数等），用于模型学习。  
目标变量（y）：二分类标签（0 = 未患病，1 = 患病），是模型需要预测的结果。  
fit_transform：先计算训练数据的均值和标准差（fit），再应用标准化（transform）。  
输出X_std：标准化后的特征矩阵，各特征均值为 0，标准差为 1。  
n_components=0.99：自动选择最少的主成分数量，使得这些主成分的方差之和占原始数据总方差的 99%。例如，若原始 8 个特征的总方差为 100，降维后 k 个主成分的方差和为 99，则 k 即为最终保留的特征数。
X_pca：降维后的特征矩阵（n_samples × k），k 由 99% 的方差保留比例决定（例如可能 k=5，表示用 5 个主成分替代原始 8 个特征）。

```python
### 对输入特征进行降维处理
from sklearn.decomposition import PCA
from sklearn import preprocessing                 # 调用标准化模块

x_cols = [col for col in df.columns if col!='Outcome']
y_col = 'Outcome'
X=df[x_cols].values
y=df[y_col].values

scaler = preprocessing.StandardScaler()           # 创建标准化对象
X_std = scaler.fit_transform(X)                   # 对数据进行标准化

pca = PCA(n_components=0.99)                      # 保留99%信息的主成分个主成分
X_pca = pca.fit(X_std).transform(X_std)           # 应用PCA进行降维

print("新的特征矩阵:\n", X_pca)
```

### 步骤6 结合相关性分析和降维后的结论，选择数据进行拆分为训练集和测试集，拆分比例设置为0.1，指定以Target的比例做分层抽样。  
test_size=0.1：测试集占比 10%（训练集 90%），平衡模型训练的充分性与评估的可靠性。
stratify=y：确保训练集和测试集的类别比例与原数据一致（例如原数据中患病样本占比 35%，训练集和测试集也应接近 35%）。  
可视化方法：使用seaborn.countplot绘制柱状图，横轴为目标变量类别（0/1），纵轴为样本数量。
数值标签：在每个柱子顶部添加具体数值（如训练集未患病样本数 = 449，患病样本数 = 242），直观展示分布比例。
验证逻辑：若训练集和测试集的类别比例与原数据一致（例如原数据中 0:1≈65:35，训练集和测试集也应接近该比例），则说明分层抽样成功。  
拆分后的结果如下

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

X = X_pca
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=42)


# 3. 可视化目标变量的分布
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.countplot(x=y_train, palette='viridis')
plt.title('训练集目标变量分布-何振')
plt.ylabel('数量')
plt.xticks([0, 1], ['未患病', '患病'])
heights = [rect.get_height() for rect in plt.gca().patches]
for i, v in enumerate(heights):
    plt.text(i, v + 0.5, f'{v}', horizontalalignment='center', verticalalignment='bottom')

plt.subplot(1, 2, 2)
sns.countplot(x=y_test, palette='viridis')
plt.title('测试集目标变量分布-何振')
plt.ylabel('数量')
plt.xticks([0, 1], ['未患病', '患病'])
heights = [rect.get_height() for rect in plt.gca().patches]
for i, v in enumerate(heights):
    plt.text(i, v + 0.5, f'{v}', horizontalalignment='center', verticalalignment='bottom')
    
plt.tight_layout()
plt.show()
```

### 步骤7 选择逻辑回归算法对拆分后的数据进行建模训练和预测，其中要求将原始模型做5折交叉验证，评估指标选择f1。
训练结果如下

```python
#引入逻辑回归和交叉验证的库
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

#引入评价指标的库
from sklearn.metrics import f1_score

# 创建逻辑回归模型
LogisticRegression_model = LogisticRegression(max_iter=1000, random_state=42)

# 使用5折交叉验证评估模型的F1分数
f1_scores = cross_val_score(LogisticRegression_model, X_train, y_train, cv=5, scoring='f1')

# 打印交叉验证的F1分数
print(f"5折交叉验证F1分数: {f1_scores}")
print(f"交叉验证平均F1分数: {f1_scores.mean():}")

# 训练模型
LogisticRegression_model.fit(X_train, y_train)

# 进行预测
y_train_pred = LogisticRegression_model.predict(X_train)
y_test_pred = LogisticRegression_model.predict(X_test)

# 计算训练集和测试集的F1分数
f1_train = f1_score(y_train, y_train_pred)
f1_test = f1_score(y_test, y_test_pred)

print(f"训练集F1分数: {f1_train:}")
print(f"测试集F1分数: {f1_test:}")

#可视化F1分数的结果
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
sns.barplot(x=['Fold 1', 'Fold 2', 'Fold 3', 'Fold 4', 'Fold 5'], y=f1_scores, palette='viridis')
plt.title('5折交叉验证F1分数-廖启宏')
plt.xlabel('交叉验证折')
plt.ylabel('F1分数')

plt.subplot(1, 2, 2)
sns.barplot(x=['训练集', '测试集'], y=[f1_train, f1_test], palette='viridis')
plt.title('训练集和测试集F1分数-廖启宏')
plt.xlabel('数据集')
plt.ylabel('F1分数')


plt.tight_layout()
plt.show()
```

### 步骤8 对逻辑回归的几个重要参数进行网格搜索
最佳参数如下

```python
from sklearn.model_selection import GridSearchCV

c_range=[0.001,0.01,0.1,1.0]
solvers = ['liblinear','lbfgs','newton-cg','sag']
max_iters=[80,100,150,200,300]
tuned_parameters= dict(solver=solvers, C=c_range,max_iter=max_iters)

lr_grid = GridSearchCV(LogisticRegression_model, param_grid=tuned_parameters, scoring='f1', cv=5)
lr_grid.fit(X_train, y_train)

# 输出最佳参数
print(f"最佳参数: {lr_grid.best_params_}")
print(f"最佳F1分数: {lr_grid.best_score_}")
```

### 步骤9 根据搜索参数，最后确认模型，进行预测。
最终训练结果如下，可以看出模型的准确率有提高，但仍然不够精确。

```python
from sklearn.metrics import classification_report, confusion_matrix

lr_best_model = lr_grid.best_estimator_

# 进行预测
y_test_pred = lr_best_model.predict(X_test)

# 计算测试集的F1分数
f1_test = f1_score(y_test, y_test_pred)

print(f"测试集F1分数: {f1_test}")
```

### 实验结果分析
#### 模型的分类报告
对测试集的真实标签与模型预测结果进行对比分析，生成包含多维度指标的评估报告。报告中会显示以下关键指标：
1. **精确率(Precision)**：预测为某类的样本中实际正确的比例（减少误诊）
2. **召回率(Recall)**：实际某类样本中被正确预测的比例（减少漏诊）
3. **F1分数**：精确率和召回率的调和平均数（综合评估）
4. **支持数(Support)**：每类的实际样本量
该报告能帮助研究者快速识别模型对不同类别的识别能力差异（例如本项目中糖尿病患者的识别效果），为后续模型优化提供量化依据。

```python
# 打印分类报告
print("\n分类报告-高远:")
print(classification_report(y_test, y_test_pred))
```

#### 模型的混淆矩阵
混淆矩阵（Confusion Matrix）是评估分类模型性能的重要工具，它以表格形式直观展示模型的预测结果与实际标签的对比情况。在这个糖尿病预测项目中，我们使用混淆矩阵来具体分析逻辑回归模型的分类表现，特别是区分"患病"和"未患病"两类时的准确性。通过矩阵中的真正例（TP）、真负例（TN）、假正例（FP）和假负例（FN），不仅能计算准确率、召回率、精确率等关键指标，也能识别模型的错误类型，比如模型漏诊（假阴性）比误诊（假阳性）更严重。此外，混淆矩阵还为优化模型提供了明确方向，比如可以通过调整分类阈值或改进特征工程来降低临床风险。

```python
# 绘制混淆矩阵
conf_matrix = confusion_matrix(y_test, y_test_pred)

plt.figure(figsize=(10, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['预测未患病', '预测患病'], yticklabels=['实际未患病', '实际患病'])
plt.title('混淆矩阵-高远')
plt.xlabel('预测')
plt.ylabel('实际')
plt.show()
```

#### 模型的ROC曲线和AUC值

ROC曲线（Receiver Operating Characteristic Curve）通过展示分类器在不同阈值下的真阳性率（True Positive Rate, TPR）与假阳性率（False Positive Rate, FPR）之间的关系来评估分类器的性能。  

AUC值（Area Under the Curve）是ROC曲线下的面积，AUC值越大，说明模型的分类效果越好。具体来说，AUC值可以解释为随机选取一个正样本和一个负样本，分类器能正确判断出哪个是正样本的概率。  

以下是模型的ROC曲线和AUC值  

```python
from sklearn.metrics import roc_curve, roc_auc_score

# 计算预测概率
y_test_prob = lr_best_model.predict_proba(X_test)[:, 1]

# 计算ROC曲线
fpr, tpr, thresholds = roc_curve(y_test, y_test_prob)

# 计算AUC
auc_score = roc_auc_score(y_test, y_test_prob)

# 绘制ROC曲线
plt.figure(figsize=(6, 6))
plt.plot(fpr, tpr, label=f'ROC曲线(AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='随机猜测')
plt.xlabel('假阳性率(FPR)')
plt.ylabel('真阳性率(TPR)')
plt.title('ROC曲线-高远')
plt.legend()
plt.show()
```

#### 模型精确率-召回率曲线
精确率-召回率曲线（Precision-Recall Curve，PR曲线）是一种用于评估分类模型性能的图形工具，特别是在类别不平衡的情况下（例如在糖尿病预测这样的场景中，正样本和负样本的数量可能差异很大）。PR曲线展示了不同阈值下的精确率（Precision）与召回率（Recall）之间的关系。  

精确率（Precision）：精确率是指在所有被预测为正的样本中，实际为正的比例。  

召回率（Recall）：召回率是指在所有实际为正的样本中，被正确识别为正的比例。 

PR曲线的作用：
- 在类别不平衡的情况下，ROC曲线可能会被误导，因为假阳性率（FPR）可能很高，即使分类器预测了很多负样本为正。而PR曲线通过关注精确率和召回率，能够更好地反映模型在正样本上的表现。
- PR曲线可以帮助用户根据具体问题的需求选择合适的分类器阈值。通过观察曲线，可以在精确率和召回率之间做出权衡，选择一个能够满足业务需求的阈值。
- PR曲线和其下的面积（AUC值）可以用于比较不同模型的性能。PR曲线下的面积（PR AUC）越接近1，表示模型的性能越好。  

从图像来看，该模型性能一般

```python
from sklearn.metrics import precision_recall_curve

# 计算精确率-召回率曲线
precision, recall, thresholds = precision_recall_curve(y_test, y_test_prob)

# 绘制曲线
plt.figure(figsize=(6, 6))
plt.plot(recall, precision, label='精确率-召回率')
plt.xlabel('召回率')
plt.ylabel('精确率')
plt.title('精确率-召回率曲线-高远')
plt.legend()
plt.show()
```

经过上面的分析，能够得出以下几个结论  

优点 ：
- AUC 值为0.80，表明模型在区分正负类样本方面表现较好。
- 在类别 0（多数类）上的精确率、召回率和 F1 分数较高，说明模型对多数类的识别能力较强。  

缺点 ：
- 在类别 1（少数类）上的精确率、召回率和 F1 分数较低，说明模型对少数类的识别能力不足。
- 精确率-召回率曲线显示模型性能一般，尤其是在高召回率下精确率下降明显。
- 错误率较高（27%），表明模型在部分样本上预测不够准确。

### 实验改进 - 更换分类算法

#### 随机森林模型

```python
from sklearn.ensemble import RandomForestClassifier

# 初始化随机森林模型
rf_model = RandomForestClassifier(random_state=42)

# 定义超参数网格
rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}

# 使用 GridSearchCV 进行网格搜索
rf_grid = GridSearchCV(rf_model, rf_params, cv=5, scoring='f1')
rf_grid.fit(X_train, y_train)

# 输出最佳 F1 分数和最佳参数
print('Random Forest Best F1 Score:', rf_grid.best_score_)
print('Random Forest Best Parameters:', rf_grid.best_params_)
```

#### 支持向量机模型

```python
import os
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# 检查数据集大小
print(X_train.shape, y_train.shape)

# 使用数据子集
X_train_small, _, y_train_small, _ = train_test_split(X_train, y_train, train_size=0.1, random_state=42)

# 初始化 SVM 模型
svc_model = SVC(random_state=42, max_iter=1000, tol=1e-3, probability=True)  # 启用概率预测

# 定义超参数网格
svc_params = {
    'C': [0.1, 1],  # 减少 C 的取值
    'kernel': ['linear']  # 只使用线性核
}

# 使用 GridSearchCV 进行网格搜索
svc_grid = GridSearchCV(
    svc_model,
    svc_params,
    cv=3,  # 减少交叉验证折数
    scoring='f1',
    n_jobs=1  # 关闭并行计算
)

# 训练模型
svc_grid.fit(X_train_small, y_train_small)  # 使用数据子集

# 输出最佳 F1 分数和最佳参数
print('SVM Best F1 Score:', svc_grid.best_score_)
print('SVM Best Parameters:', svc_grid.best_params_)
```

#### K近邻模型

```python
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier()
knn_params = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance']
}
knn_grid = GridSearchCV(knn_model, knn_params, cv=5, scoring='f1')
knn_grid.fit(X_train, y_train)
print('KNN Best F1 Score:', knn_grid.best_score_)
print('KNN Best Parameters:', knn_grid.best_params_)
```

#### XGBoost模型

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def plot_confusion_matrix(y_true, y_pred, title, ax):
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['预测未患病', '预测患病'],
                yticklabels=['实际未患病', '实际患病'],
                ax=ax)
    ax.set_xlabel('预测')
    ax.set_ylabel('实际')
    ax.set_title(title)
# 创建一个3x2的子图布局
fig, axes = plt.subplots(3, 2, figsize=(12, 18))

# 为每个模型绘制混淆矩阵
plot_confusion_matrix(y_test, lr_grid.best_estimator_.predict(X_test), '逻辑回归混淆矩阵-马瑞', axes[0, 0])
plot_confusion_matrix(y_test, rf_grid.best_estimator_.predict(X_test), '随机森林混淆矩阵-马瑞', axes[0, 1])
plot_confusion_matrix(y_test, svc_grid.best_estimator_.predict(X_test), '支持向量机混淆矩阵-马瑞', axes[1, 0])
plot_confusion_matrix(y_test, knn_grid.best_estimator_.predict(X_test), 'K近邻混淆矩阵-马瑞', axes[1, 1])
plot_confusion_matrix(y_test, xgb_grid.best_estimator_.predict(X_test), 'XGBoost混淆矩阵-马瑞', axes[2, 0])

# 隐藏最后一个空的子图
axes[2, 1].axis('off')

# 调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()
```

#### 绘制多个模型的 ROC 曲线

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# 定义函数绘制 ROC 曲线
def plot_roc_curve_multiple(y_true, models, model_names, X_test):
    plt.figure(figsize=(8, 6))
    for model, name in zip(models, model_names):
        y_pred_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_true, y_pred_prob)
        auc_score = roc_auc_score(y_true, y_pred_prob)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {auc_score:.2f})')
    
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='随机猜测')
    plt.xlabel('假阳性率(FPR)')
    plt.ylabel('真阳性率(TPR)')
    plt.title('多个模型的ROC曲线对比图-马瑞')
    plt.legend()
    plt.show()

# 模型列表
models = [
    lr_grid.best_estimator_,
    rf_grid.best_estimator_,
    svc_grid.best_estimator_,
    knn_grid.best_estimator_,
    xgb_grid.best_estimator_
]
model_names = ['逻辑回归', '随机森林', '支持向量机', 'K近邻', 'XGBoost']

# 绘制 ROC 曲线
plot_roc_curve_multiple(y_test, models, model_names, X_test)
```

#### 多个模型的分类报告

```python
from sklearn.metrics import classification_report
from prettytable import PrettyTable

def generate_classification_report_table(y_true, models, model_names, X_test):
    reports = []
    for model, name in zip(models, model_names):
        y_pred = model.predict(X_test)
        report = classification_report(y_true, y_pred, output_dict=True)
        report_df = pd.DataFrame(report).transpose()
        report_df['Model'] = name
        reports.append(report_df)
    
    # 合并所有报告
    final_report = pd.concat(reports)
    
    # 调整格式
    final_report = final_report.reset_index().set_index(['Model', 'index'])  # 将 Model 和 index 作为多级索引
    final_report = final_report[['precision', 'recall', 'f1-score', 'support']]  # 只保留关键指标
    final_report = final_report.round(2)  # 将数值格式化为两位小数
    
    # 过滤掉 accuracy、macro avg 和 weighted avg
    final_report = final_report.loc[(slice(None), ['0', '1']), :]
    
    return final_report

# 生成分类报告表格
report_table = generate_classification_report_table(y_test, models, model_names, X_test)

# 创建两个表格，分别对比 Class 0 和 Class 1
table_class_0 = PrettyTable()
table_class_0.field_names = ['Model', 'Class', 'Precision', 'Recall', 'F1-Score', 'Support']

table_class_1 = PrettyTable()
table_class_1.field_names = ['Model', 'Class', 'Precision', 'Recall', 'F1-Score', 'Support']

# 填充表格
for (model, cls), row in report_table.iterrows():
    if cls == '0':
        table_class_0.add_row([model, cls, row['precision'], row['recall'], row['f1-score'], row['support']])
    elif cls == '1':
        table_class_1.add_row([model, cls, row['precision'], row['recall'], row['f1-score'], row['support']])

# 打印表格
print("多个模型的分类报告-马瑞")
print("Class 0 Comparison:")
print(table_class_0)
print("\nClass 1 Comparison:")
print(table_class_1)
```

#### 绘制多个模型的精确率-召回率曲线

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# 定义函数绘制精确率-召回率曲线
def plot_precision_recall_curve_multiple(y_true, models, model_names, X_test):
    plt.figure(figsize=(8, 6))
    for model, name in zip(models, model_names):
        y_pred_prob = model.predict_proba(X_test)[:, 1]
        precision, recall, _ = precision_recall_curve(y_true, y_pred_prob)
        plt.plot(recall, precision, label=f'{name}')
    
    plt.xlabel('召回率')
    plt.ylabel('精确率')
    plt.title('多个模型的精确率-召回率曲线-马瑞')
    plt.legend()
    plt.show()

# 绘制精确率-召回率曲线
plot_precision_recall_curve_multiple(y_test, models, model_names, X_test)
```

### 实验改进 - 随机欠采样
从过采样的结果可以看出，模型的效果有了改善但仍然存在提升空间，尤其是XGBoost模型，过采样后模型效果提升不明显。因此我们决定继续探索其他方法，尝试提升模型效果。  

我们选择尝试随机欠采样法，欠采样是从多数类中随机选择部分样本，使得两类样本数量更加均衡。这样可以使得模型在训练过程中不会过度偏向于多数类（未患病），从而提高对少数类（患病）的预测能力。  

这里随机删除了测试集中的未患病样本，使两个样本的个数都为241。

```python
from imblearn.under_sampling import RandomUnderSampler

# 初始化欠采样
undersampler = RandomUnderSampler(random_state=42)

# 对训练数据进行欠采样
X_train_undersampled, y_train_undersampled = undersampler.fit_resample(X_train, y_train)

# 查看欠采样后的类别分布
print('欠采样后的类别分布:', Counter(y_train_undersampled))
```

#### 训练模型
使用欠采样处理后的训练集，训练前面用到的5个模型。

```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif

# 标准化特征
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_undersampled)
X_test_scaled = scaler.transform(X_test)

# 特征选择
selector = SelectKBest(score_func=f_classif, k=8)
X_train_selected = selector.fit_transform(X_train_scaled, y_train_undersampled)
X_test_selected = selector.transform(X_test_scaled)

# 训练逻辑回归模型
lr_US_model = LogisticRegression(random_state=42, class_weight='balanced')
lr_US_model.fit(X_train_selected, y_train_undersampled)
y_US_pred_lr = lr_US_model.predict(X_test_selected)

# 训练随机森林模型
rf_US_model = RandomForestClassifier(random_state=42, class_weight='balanced')
rf_US_model.fit(X_train_selected, y_train_undersampled)
y_US_pred_rf = rf_US_model.predict(X_test_selected)

# 训练支持向量机模型
svc_US_model = SVC(random_state=42, class_weight='balanced', probability=True)
svc_US_model.fit(X_train_selected, y_train_undersampled)
y_pred_US_svc = svc_US_model.predict(X_test_selected)

# 训练K近邻模型
knn_US_model = KNeighborsClassifier(weights='distance')
knn_US_model.fit(X_train_selected, y_train_undersampled)
y_US_pred_knn = knn_US_model.predict(X_test_selected)

# 训练XGBoost模型
xgb_US_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss', scale_pos_weight=500/268)
xgb_US_model.fit(X_train_selected, y_train_undersampled)
y_US_pred_xgb = xgb_US_model.predict(X_test_selected)
```

#### 混淆矩阵
从混淆矩阵中可以看出，所有模型在未患病样本中的精确率都有提升，召回率有所降低；而在患病样本中的精确率都有降低但召回率有所提升。与未处理过的数据集相比，模型预测未患病人群的准确率有提升，但在患病人群中准确率有所降低。

```python
undersalample_models = [
    lr_US_model,
    rf_US_model,
    svc_US_model,
    knn_US_model,
    xgb_US_model
]
models_names = ['逻辑回归', '随机森林', '支持向量机', 'K近邻', 'XGBoost']

# 创建3x2的子图布局
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(8, 12))

# 遍历所有模型生成混淆矩阵
for i, model in enumerate(undersalample_models):
    # 进行预测
    y_test_pred = model.predict(X_test)
    
    # 计算混淆矩阵
    conf_matrix = confusion_matrix(y_test, y_test_pred)
    
    # 获取子图的行和列位置
    row = i // 2
    col = i % 2
    
    # 绘制混淆矩阵
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['预测未患病', '预测患病'], yticklabels=['实际未患病', '实际患病'], ax=axes[row, col])
    axes[row, col].set_title(f'{models_names[i]}混淆矩阵-魏裕堃')
    axes[row, col].set_xlabel('预测')
    axes[row, col].set_ylabel('实际')

axes[2, 1].axis('off')
# 调整子图之间的间距
plt.tight_layout()
plt.show()
```

#### ROC曲线对比图
从ROC曲线和AUC值来看，与未处理过的数据集相比，逻辑回归、支持向量机两个模型的效果有较小的提升，而其他三个模型的效果有所降低。

```python
# 创建一个图和轴
plt.figure(figsize=(6, 6))

# 遍历所有模型生成ROC曲线
for model, name in zip(undersalample_models, models_names):
    # 如果模型是逻辑回归或支持向量机，它们可以直接输出概率
    if isinstance(model, (LogisticRegression, SVC)):
        y_prob = model.predict_proba(X_test)[:, 1]
    elif isinstance(model, RandomForestClassifier):
        y_prob = model.predict_proba(X_test)[:, 1]
    elif isinstance(model, KNeighborsClassifier):
        y_prob = model.predict_proba(X_test)[:, 1]
    elif isinstance(model, XGBClassifier):
        y_prob = model.predict_proba(X_test)[:, 1]
    else:
        raise ValueError("未知的模型类型")
    
    # 计算ROC曲线和AUC值
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    roc_auc = roc_auc_score(y_test, y_prob)
    
    # 绘制ROC曲线
    plt.plot(fpr, tpr, lw=2, label=f'{name} (AUC = {roc_auc:.2f})')

# 绘制对角线
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')

# 设置图的属性
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('假阳性率 (FPR)')
plt.ylabel('真阳性率 (TPR)')
plt.title('欠采样后多个模型的ROC曲线-魏裕堃')
plt.legend(loc='lower right')

# 显示图形
plt.show()
```

#### 精确率-召回率曲线对比图
从精确率-召回率曲线来看，与未处理过的数据集相比，模型的效果大体上相同。

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, average_precision_score


# 创建一个图和轴
plt.figure(figsize=(6, 6))

# 遍历所有模型生成PR曲线
for model, name in zip(undersalample_models, models_names):
    # 如果模型是逻辑回归或支持向量机，它们可以直接输出概率
    if isinstance(model, (LogisticRegression, SVC)):
        y_prob = model.predict_proba(X_test)[:, 1]
    elif isinstance(model, RandomForestClassifier):
        y_prob = model.predict_proba(X_test)[:, 1]
    elif isinstance(model, KNeighborsClassifier):
        y_prob = model.predict_proba(X_test)[:, 1]
    elif isinstance(model, XGBClassifier):
        y_prob = model.predict_proba(X_test)[:, 1]
    else:
        raise ValueError("未知的模型类型")
    
    # 计算PR曲线和平均精度
    precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
    avg_precision = average_precision_score(y_test, y_prob)
    
    # 绘制PR曲线
    plt.plot(recall, precision, lw=2, label=f'{name} (AP = {avg_precision:.2f})')

# 设置图的属性
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('召回率')
plt.ylabel('精确率')
plt.title('欠采样后模型的精确率-召回率曲线对比-魏裕堃')
plt.legend(loc='best')

# 显示图形
plt.show()
```

#### 分类报告
通过分类报告给出的具体数据，我们可以得出以下结论：
- 对于类别1（患病），模型召回率大幅提升，更擅长识别患病样本，减少漏诊风险。但精确率下降，有着更高的假阳性率。
- 对于类别0（未患病），模型召回率降低，对未患病样本的识别能力下降，可能导致过度诊断。

- 若任务更关注患病样本的检出率（如疾病筛查），欠采样后的模型更合适；若需平衡精确率和召回率，则需要探索其他方法。

```python
from sklearn.metrics import classification_report
import pandas as pd

# 初始化存储容器
class0_data = []
class1_data = []

# 遍历所有模型生成报告
for model in undersalample_models:
    # 根据模型类型预测测试集的类别
    if isinstance(model, (LogisticRegression, RandomForestClassifier, SVC, KNeighborsClassifier)):
        y_pred = model.predict(X_test_selected)
    elif isinstance(model, XGBClassifier):
        y_pred = model.predict(X_test_selected)
    else:
        raise ValueError("未知的模型类型")
    
    # 生成分类报告字典
    report = classification_report(y_test, y_pred, output_dict=True)
    
    # 提取类别0指标
    class0 = report['0']
    class0_data.append({
        'Model': model.__class__.__name__,
        'Precision': round(class0['precision'], 3),
        'Recall': round(class0['recall'], 3),
        'F1-Score': round(class0['f1-score'], 3),
        'Support': int(class0['support'])
    })
    
    # 提取类别1指标
    class1 = report['1']
    class1_data.append({
        'Model': model.__class__.__name__,
        'Precision': round(class1['precision'], 3),
        'Recall': round(class1['recall'], 3),
        'F1-Score': round(class1['f1-score'], 3),
        'Support': int(class1['support'])
    })

# 创建DataFrame并打印结果
df_class0 = pd.DataFrame(class0_data)
df_class1 = pd.DataFrame(class1_data)

print("分类报告（欠采样）-魏裕堃".center(50))
print("▂" * 40)
print("类别 0 - 未患病".center(50))
print(df_class0.to_string(index=False))
print("\n" + "▂" * 40)
print("类别 1 - 患病".center(50))
print(df_class1.to_string(index=False))
print("▂" * 40)
```

### 展望
由上面的分析可以得出，欠采样通过减少多数类样本数量，显著提升了模型对少数类（患病）的识别能力，但代价是多数类（未患病）的性能下降。需根据实际需求（如医疗场景中更关注召回率）选择模型，并进一步通过阈值调整或集成方法优化性能。   

对于进一步改进方向，可以考虑通过混合采样方法如‌SMOTE-Tomek、SMOTE-ENN或是更换数据集的方法来进一步提升模型的性能。
